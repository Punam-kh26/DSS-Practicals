```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x=np.arange(1,11)
```

```python
x
```

```python
y=x*2
```

```python
y
```

# Line Chart

```python
plt.plot(x,y)
plt.title("Line Chart")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

# Bar Chart

```python
plt.bar(x,y,color=["#9B5DE0","#D78FEE"],edgecolor="black")
plt.title("Bar Chart")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()
```

# Scatter Plot

```python
a=(1,4,2,6,8,2,5,9,0,6)
b=(23,44,12,65,77,44,13,54,98,65)
plt.scatter(a,b ,color="#9B5DE0",s=150)
plt.title("Scatter Chart")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()
```

# Pie Chart

```python
students = [25, 30, 20, 15]
batches = ['Batch 1', 'Batch 2', 'Batch 3', 'Batch 4']
plt.figure(figsize=(6,6))
plt.pie(students, labels=batches, autopct='%1.1f%%', startangle=90,
colors=['#A8DADC', '#457B9D', '#F4A261', '#E76F51'])
plt.title("Student Distribution in Batches 14")
plt.axis('equal')
# Show chart
plt.show()
```

# Histogram

```python
a = [3, 5, 1, 7, 2, 8, 6, 9, 4, 5,
2, 7, 8, 3, 6, 9, 5, 4, 7, 2,
1, 8, 6, 5, 9, 3, 4, 7, 2, 8]
# Create histogram
plt.hist(a, bins=8, color='skyblue' , edgecolor="black")
# Labels and title
plt.title("Histogram of Data")
plt.xlabel("Value Range")
plt.ylabel("Frequency")
# Add grid
plt.grid(axis='y', linestyle='--', alpha=0.7)
# Show plot
plt.show()
```
