```python
import numpy as np
x=np.array([1,2,3,4,5,6,7,2,6,2,1,4,2,2,6])
```

```python
print(np.mean(x))
```

```python
from scipy import stats
```

```python
print(stats.mode(x))
```

```python
print(np.std(x))
```

```python
print("Standard Deviation : ",std_dev)
```
